###  Python Example

The following example is meant to sketch an outline for the typical solution to a differential equation in Python.  First, we can approximate the solution using Euler's method.  Next, we can plot the solution including a slope field of points.  Finally, we can solve the IVP exactly using `sympy`.

```python
#define differential equation 
def df(x, y):
    return x + y
```

```python
#implement Euler's method with step size dt = 0.1
#initial condition f(0) = 1
dt = 0.1
x = [0]
y = [1]
for i in range(10):
    next = y[i] + dt*df(x[i], y[i])
    y.append(next)
    x.append(x[i] + dt)
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
nx, ny = (len(x), len(y))

X, Y = np.meshgrid(np.arange(0, 1, 0.1), np.arange(0, 14, 1))
U, V = df(X, Y)[:1], df(X, Y)[1:2]
```
