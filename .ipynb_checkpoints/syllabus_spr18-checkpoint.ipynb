### Calculus with a Computer


**Overview**

This class is an introduction to the foundational ideas in Calculus.  Integration, differentiation, and differential equations are introduced to familiarize students with the discipline and to show it's applications to a wide variety of problems.  We will use the computer and introduce the Python computer language to complete our work. 

#### Learning Objectives

- Understand the mathematical concepts and history of Integration
- Use integration to solve problems
- Understand the mathematical concept and history of Differentiation
- Use differentiation to solve problems
- Connect Integration and Differentiation through Differential Equations and explore the history of these problems
- Use differential equations to solve problems

### Structure of Course

We will work on each of the topics listed for 1  class.  This will involve viewing instructional videos in advance of class, and considering a few brief problems.  You will bring your ideas to class to complete a problem set with peers that will be submitted and graded weekly.  In addition to these problems, you will complete a short project for each section that goes deeper in a particular area as discussed in class.  These will result in a short group video presentation.

At the end of the course, we will complete short review problem sets and work on a final project that builds out an earlier project investigation.  This will result in an individual video presentation and paper.

Upon completion of the course, students will have a portfolio of problem sets, group projects, and individual research work.
