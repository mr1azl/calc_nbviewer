```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np
```

```python
output_notebook()
```

```python
from bokeh.models import Quad

x = np.linspace(0,4,100)
y = x**2
p = figure(title = "Approximating Areas", tools = "tap")
p.line(x, y, line_width = 3)
p.quad(top = [1,4,9,16], bottom=[0, 0, 0, 0], left=[0, 1, 2, 3], 
       right=[1, 2, 3, 4], selection_color = 'orange',
       nonselection_fill_alpha = 0.2,
       nonselection_fill_color = "blue",
       nonselection_line_color = "black",
       line_color = 'black', alpha = 0.3)
p.quad(top = [0, 1, 4, 9], bottom=[0, 0, 0, 0], left=[0, 1, 2, 3], right=[1, 2, 3, 4], 
       selection_color = 'firebrick',
       nonselection_fill_alpha = 0.2,
       nonselection_fill_color = "grey",
       nonselection_line_color = "black",
       line_color = 'black', alpha = 0.3)





```

```python
output_notebook()
show(p)
```

```python
from bokeh.models import HoverTool

x = np.linspace(0,4,100)
y = x**2
tools = ["tap", '']
p = figure(title = "Approximating Areas", tools = tools)

p.title.text_color = "grey"
p.title.text_font = "courier"
p.title.text_font_style = "bold"
p.title.text_font_size = '14pt'


p.line(x, y, line_width = 3)
cr = p.quad(top = [1,4,9,16], bottom=[0, 0, 0, 0], left=[0, 1, 2, 3], 
       right=[1, 2, 3, 4], hover_color = 'orange',
       hover_fill_alpha = 0.2,
       hover_fill_color = "blue",
       hover_line_color = "blue",
       line_color = 'black', alpha = 0.3)


p.add_tools(HoverTool(tooltips=None, renderers=[cr]))

```

```python
output_notebook()
show(p)
```
