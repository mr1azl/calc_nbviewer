```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact, widgets, fixed
```

```python
def simple_pop(N=100, x0 = 1000, c = 10):
    x = np.zeros(N)
    x[0] = x0
    for n in range(N-1):
        x[n+1] = x[n] + c
    return x

x = simple_pop(N=10, x0 = 200)
```

```python
interact(simple_pop, N = fixed(10));
```

```python
def simple_plot(i,x):
    plt.figure(figsize=(9,6))
    plt.plot(x[:i],'-ok', linewidth=2)
    plt.ylim(0,2000); plt.xlim(0,len(x))
    plt.xlabel('Year'); plt.ylabel('Population')
    plt.xlabel('Year')
    plt.ylabel('Population')
    plt.title('Simple Fish Population', loc = 'left')
    plt.xticks([i for i in np.arange(1,20,1)])
```

```python
x = simple_pop(N=20)
simple_plot(len(x), x)
```

```python
interact(simple_plot, i=10, x=fixed(x))
```
