### Qualitative Description

Despite the potentially obvious solutions from above in terms of trigonometric functions, we want to examine a qualitative solution to the problem.  Just as in class, we can use a phase plane to first examine the differential equation.  In order to do so, we can rewrite the differential equation recognizing that if $x$ is a displacement, then it's derivative is a velocity.  Further, we can use these pieces of information to sketch the phase plane by considering a few starting places $x$ and following its movement using the velocity to draw our isoclines.  

To start, we rewrite the original differential equation in terms of displacement $x$ and velocity $v$ as follows:

$$\dot{x} = v$$
$$\dot{v} = \frac{-k}{m} x$$

Now, we have a grid of points at coordinate pairs $(x, v)$.  Using the grid below, we can sketch some arrows to get a feel for what's happening


```python
x = np.arange(-3,4)
x, y = np.meshgrid(x, x)
v = np.arange(-2,3)

plt.figure("Empty Grid")
plt.plot(x, y, 'o', color = 'black')
plt.axhline()
plt.axvline()
```

$$\omega^2x^2 + v^2 = C$$

where $C \geq 0$ is some constant.  This should line up with the familiar family of ellipses from high school mathematics.
