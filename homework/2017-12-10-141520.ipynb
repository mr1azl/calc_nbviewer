# Methods
For the variables, we did four pair. For all four sets of data, we each did one scatter plot, one scatter plot with trend line, two box- plots for each variable and two histograms for each variable. We did scatterplot to see the relationship between the two variables, and the trend line helps us to see the relationship more clearly and identify if it is a strong or weak relationship. We did box-plots because it can help us to see the range, the median of the data and since it is separated by a categorical variable, we can discover the relationship between the continuous variable and the categorical variable. We did the histogram because it allows us to see the maximum and minimum for each variable. 


# First variable 
The first pair is the male earnings and female earnings. We collected the data from the HDI dataset. The data could be conducted through surveys, social research or statistic report from authentic companies. This pair of variable is important because it can show the difference between female and male. The equality between female and male is a very controversial topic right now, it is crucial for us to find the gap through data comparison. For the outside source, we found that there is an EqualPay Act since 1963. Since then, the United States labor law abandoned the wage gap between male and female. But even with the act, we can still see disparity on salaries between male and female nowadays. 


# Hypothesis
We assume that male earns more than female do. We do not think there will be an obvious relationship for scatterplot for that they might not influence each other directly, it is just a phenomenon. 

<font color = "red"> Okay, but how else might you compare these?  Why would there be a relationship between them?  What about mean, median, mode or descriptive comparisons from one group to the next?</font>


First pair of variables:We made six graphs.

```r
ggplot(HDI) + 
  geom_point(mapping = aes(x = Earnings_M , y = Earnings_F ))
```

There is not a strong relationship, the plots are very scattered and it is hard to see the relationship.

```r
ggplot(data = HDI, mapping = aes(x = Earnings_M, y = Earnings_F)) + 
  geom_point() +
    geom_smooth()
```

With the trend line, we are able to see the relationship more clearly. At the beginning, the more the female earns, the less the male earns, but as the graph develop, we can see that the more female earns, the more male earns. And despite the female salary is on the rise, male earns more than female. 

<font color = "red"> Okay, great.  The two plots below would be great either side by side or with the groups compared within one another.  For example, can you show each region where you've divided it by sex?  One boxy for each next to each other so you can compare the distribution of female earnings versus male earnings?</font>

```r
b1  <- ggplot(HDI) +
  geom_boxplot(mapping = aes(x = Region, y = Earnings_F )) +
    theme(axis.text.x = element_text(angle=70, vjust=0.5))
```

The earning of female is catagorized by region. We can see that female from the midwest earned the least and female from the South earns the most.

```r
b2  <- ggplot(HDI) +
  geom_boxplot(mapping = aes(x = Region, y = Earnings_M )) +
    theme(axis.text.x = element_text(angle=70, vjust=0.5))

plot_grid(b1, b2)
```

The earning of male is catagorized by region. We can see that male from the NorthEast earns the most and far exceed other regions. And the male from the South earns the least based on the median. However, the maximum wage for male is the least from the Midwest. And there are only few outliers. 

<font color = "red"> Yes, good.  Again, I want you to use these histograms to compare to each other.  </font>

```r
h1  <-  ggplot(data = HDI) + 
  geom_histogram(mapping = aes(x = Earnings_M), bins = 20)
```

From the histogram, we can see that most of the male people earns annually around $32000. There are very few that earns below $30000 or exceed $37500. 

```r
h2  <- ggplot(data = HDI) + 
  geom_histogram(mapping = aes(x = Earnings_F), bins = 20)
```

```r
plot_grid(h1, h2)
```

From the histogram, we can see that most of the female earns annully around $22000. There are very few that earn below $20000 or exceed $30000. 

```r
summary(lm(Earnings_M ~ Earnings_F, data=HDI))
```
