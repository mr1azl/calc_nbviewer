# Calculus, Culture, Computers

This class aims to introduce the foundational concepts of the Calculus.  We do so by examining the historical trajectory of the concepts and investigate a number of the primary sources.  Also, we aim to investigate a sample of ways people have then used these ideas to solve problems in disciplines outside of mathematics.  Here, we will use Jupyter notebooks and the Python computer language to perform computations and present our results.

**GOALS**:

- Introduce integration, differentiation, and differential equations

- Investigate historical roots of Calculus

- Use computer to apply Calculus to variety of fields

## Section I: Summation and the Definite Integral

The first section deals with what we understand as the initial problem of the Calculus -- integration.  The infinite process has a long history.  In this section we will begin to perform an infinite number of operations on numbers and connect this to a strategy for finding areas and volumes.  We also investigate how others have used integration to solve problems in biology, economics, physics, and probability and statistics.

### Week 1

- **0.0.1: Monday 8.28**: *Introduction to Class*.  Today we will discuss the big picture of Calculus, its history and contributions to culture and society.  Next, we outline the schedule and structure for the course.  Finally, we turn to the computers for a brief introduction to plotting with Python.


- **0.0.2: Wednesday 8.30**: *Modeling with Sequences and Functions*. Today we explore some important families of functions that will continue to arise in our work.  Linear, Quadratic, Exponential, and Trigonometric functions defined sequentially with loops and in closed form with functions are represented and discussed.  We relate these situations to modeling various phenomena.


### Week 2

- **0.0.3: Wednesday 9.6**: *The Infinite Process*.  This notebook begins our exploration of the infinite process with the Babylonian Square root algorithm, Archimedes exhaustion of a circle, and Zeno's Paradox. These problems demonstrate early uses of the infinite process so important to Calculus.  Also, these examples demonstrate some problems with number, quantity, and representation.



### Week 3

- **0.0.4: Monday 9.11**: *Summations and Integration*. Now, we connect the infinte process to problems involving sums and powers of integers.  These problems are solved in Python numerically and symbolically. Finally, we introduce the notation of the definite integral that we will use throughout the class.



- **0.0.5: Wednesday 9.13**: *Area and Summation*.  We put our work with summations to use in solving area problems.  The general approach of using rectangles to approximate non-regular areas connects the earlier work with summations to the area problem.  



### Week 4

- **0.0.6: Monday 9.18**: *Cardiac Output*. Summations and Area approximations are applied to finding the cardiac output of an organism.   



- **0.0.7: Wednesday 9.20**: *The Definite Integral*. Based on our approximation strategies, we determine a way to define the exact area under a curve with the definite integral.  Notation and implementation with Python are both discussed.



### Week 5

- **0.0.8: Monday 9.25**: *Consumer Surplus and the Definite Integral*. We apply our definition of the definite integral to solve problems from economics involving supply, demand, equilibrium, revenue, and consumer surplus.  



- **0.0.9: Wednesday 9.27**: *Area Between Curves and the Gini Index*.  Building on the work with Consumer Surplus, we extend the use of definite integrals to determine the area between two curves.  Finally, we examine how this can be applied to problems of equal distribution of wealth with the Gini Index.
