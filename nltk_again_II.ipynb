```python
url = "https://en.wikipedia.org/wiki/Mathematics_education"
```

```python
import requests
```

```python
r = requests.get(url)
```

```python
type(r)
```

```python
html = r.text
```

```python
print(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, "html5lib").get_text()
```

```python
from nltk import word_tokenize
tokens = word_tokenize(soup)
```

```python
import re
words = [w for w in tokens if re.search('^[A-Za-z]{3,}', w)]
```
